# 数据选择

## 1.列选择

- 选择某一列
- 选择某几列

**方法**
- 在Python中我们要想获取某列只需要在表df后面的方括号中指明要选择的列名即可。
- 如果是一列，则只需要传入一个列名；
- 如果是同时选择多列，则传入多个列名即可，多个列名用一个list存起来。
- 在Python中我们把这种通过传入列名选择数据的方式称为普通索引。

### 普通索引（传入列名）

````python
# 导入依赖
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")

# 输出行索引index和列索引columns
print('列索引:{}'.format(df.columns))
print('行索引:{}'.format(df.index))

#  选择某一列
# 在Python中我们把这种通过传入列名选择数据的方式称为普通索引
df_col = df["CustomerID"]
print('选中CustomerID列为：\n{}'.format(df_col))
````

### 位置索引（传入位置）

- 可以传入具体列的位置，即第几列，对数据进行选取
- 通过传入位置来获取数据时需要用到iloc方法。

````python
# 导入依赖
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")

# 获取第 1 列和第 3 列的值

# iloc 后的方括号中逗号之前的部分表示要获取的行的位置，只输入一个冒号，不输入任何数值表示获取所有的行；
# 逗号之后的方括号表示要获取的列的位置，列的位置同样是也是从0开始计数。
df_1_3 = df.iloc[:, [0,2]]
print("第1、3列的值为：\n{}".format(df_1_3))

````

### 切片索引

- 我们把这种通过传入一个位置区间来获取数据的方式称为切片索引

````python
# 导入依赖
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")

# 获取1~3列的连续列
df_13 = df.iloc[:, 0:3]

print("第1~3列的值为：\n{}".format(df_13))
````

## 2.行选择

- 普通索引，即传入具体行索引的名称，需要用到loc方法；

- 另一种是位置索引，即传入具体的行数，需要用到iloc方法。


### 普通索引（loc）

即传入具体行索引的名称，需要用到loc方法

````python
# 导入依赖
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")

# 选择第一行
df_row_1 = df.loc[0] #loc后面给的是索引的名称
print("第1行的数据为：\n{}".format(df_row_1))

# 选择第1、3行
df_row_1_3 = df.loc[[0,2], :]
print("第1、3行的数据为：\n{}".format(df_row_1_3))


````

### 位置索引（iloc）

- 单选
- 选择连续的某几行

````python
# 导入依赖
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")
# 利用iloc方法
# 选择第一行
df_row_1 = df.iloc[0] #iloc给的是行号
print("第1行的数据为：\n{}".format(df_row_1))

# 选择第一行和第二行
df_row_1_2 = df.iloc[[0,1], :]
print("第1、2行的数据为：\n{}".format(df_row_1_2))

# 选择连续的某几行(两种表达)
# df_row_1_4 = df.iloc[0:4, :] 
df_row_1_4 = df.iloc[0:4]
print("第1~4行的数据为：\n{}".format(df_row_1_4))
````

### 选择满足条件的行

- 获取某一列时，获取的是这一列的所有行，我们还可以只筛选出这一列中满足条件的值。

- 我们直接在表名后面指明哪列要满足什么条件，就可以把满足条件的数据筛选出来。


````python
# 导入依赖
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")

# 选择年龄小于40岁的行
df_age = df[df["Age"] < 40]
print("年龄小于40岁的行为：\n{}".format(df_age))

# 选择年龄小于40岁且大于20岁的行(这里是一个括号包括了表达式)
df_age_between = df[(df["Age"] < 40) & (df["Age"] > 20)]
print("年龄小于40岁且大于20岁的行为：\n{}".format(df_age_between))
````

### 行列同时选择

- 实际业务中我们也会用到行、列同时选择

- 所谓的行、列同时选择就是选择出行和列的相交部分。

**方法**

- 普通索引 + 普通索引： 通过同时传入行和列的索引名称进行数据选择，需要用到loc方法。

- 位置索引+位置索引： 通过同时传入行、列索引的位置来获取数据，需要用到iloc方法。

- 布尔索引+普通索引： 先对表进行布尔索引选择行，然后通过普通索引选择列

- 切片索引+切片索引： 通过同时传入行、列索引的位置区间进行数据选择。

- 切片索引+普通索引： 如果是普通索引，就直接传入行或列名，用loc方法即可；如果是切片索引，也就是传入行或列的位置区间，要用 iloc 方法。如果是切片索引+普通索引，也就是行（列）用切片索引，列（行）用普通索引，这种交叉索引要用ix方法。


````python
# 普通索引 + 普通索引： 通过同时传入行和列的索引名称进行数据选择，需要用到loc方法。
# 导入依赖
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")

#  获取第一行、 第三行 和 第一列、 第三列数据
# 用 loc 方法传入行列名称
df_13_13 = df.loc[[0, 2], ["CustomerID", "Age"]]
print(df_13_13)
# loc方法中的第一对方括号表示行索引的选择，传入行索引名称；loc方法中的第二对方括号表示列索引的选择，传入列索引名称。
````

````python
# 位置索引+位置索引是通过同时传入行、列索引的位置来获取数据，需要用到iloc方法。
# 导入依赖
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")
#  获取第一行、 第三行 和 第一列、 第三列数据
df_13_13 = df.iloc[[0, 2], [0, 2]]
print(df_13_13)
# 在iloc方法中的第一对方括号表示行索引的选择，传入要选择行索引的位置；第二对方括号表示
# 列索引的选择，传入要选择列索引的位置。行和列索引的位置都是从0开始计数。
````

````python
# 布尔索引+普通索引是先对表进行布尔索引选择行，然后通过普通索引选择列。

# 导入依赖
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")
#  获取年龄小于40大于20的行和 第一列、 第三列数据
df_bool = df[(df["Age"] < 40) & (df["Age"] > 20)][["CustomerID", "Age"]]
print(df_bool)

# 先通过布尔索引选择出年龄小于200的所有行，然后通过普通索引选择订单编号和年龄这两列。
````

````python
# 切片索引+切片索引是通过同时传入行、列索引的位置区间进行数据选择。

# 导入依赖
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")

# 选择1~3行， 1~3列
df_slice = df.iloc[0:4, 0:4]
print(df_slice)
````

## 1.数值替换

- 数值替换就是将数值A替换成B，可以用在异常值替换处理、缺失值填充

- 主要有一对一替换、多对一替换、多对多替换三种替换方法。

### 一对一替换

- 一对一替换是将某一块区域中的一个值全部替换成另一个值。

**Excel**:

- 在Excel中对某个值进行替换，首先要把待替换的区域选中，如果只是替换某一列中的值，只需要选中这一列即可；如果要在一片区域中进行替换，那么拖动鼠标选中这一片区域。然后依次单击编辑菜单栏中的查找和选择>替换选项即可调出替换界面。使用快捷键Ctrl+H也可以调出替换界面。

- 在Python中对某个值进行替换利用的是replace()方法，replace(A,B)表示将A替换成B。

````python
# 切片索引+切片索引是通过同时传入行、列索引的位置区间进行数据选择。

# 导入依赖
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")

# 将 30 岁的年龄替换成 33 岁
df['Age'].replace(30, 33)

# 对全表中的缺失值进行替换，这个时候replace()方法就相当于fillna()方法了。
# np.NaN是Python中对缺失值的一种表示方法。
df.replace(np.NAN, 0)
````

### 多对一替换

- 多对一替换就是把一块区域中的多个值替换成某一个值，已知现在有三个异常年龄（240、260、280）需要把这三个年龄都替换成正常范围年龄的平均值33

**Excel实现**

- 在Excel中需要借助if函数来实现多对一替换。

**Python实现**

- 同样也是利用 replace()方法， replace([A,B],C)表示将A、B替换成C。

````python
# 导入依赖
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")

# 将 30 岁的年龄替换成 33 岁
df['Age'].replace([19, 35, 35], 33)
````

### 多对多替换

- 多对多替换其实就是某个区域中多个一对一的替换。

**Excel**

- 需要借助函数，且需要多个if嵌套语句来实现

**Python**

- 借助replace()方法，将替换值与待替换值用字典的形式表示，replace({"A":"a","B":"b"}表示用a替换A，用b替换B。


````python
# 导入依赖
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")

df["Age"].replace({33: 20, 23: 21, 22: 19})
````

## 2.数值排序

数值排序是按照具体数值的大小进行排序：

- 升序
- 降序



### 按照一列数值进行排序

- 按照一列数值进行排序就是整个数据表都以某一列为准，进行升序或降序排列。

- 在 Python 中我们若想按照某列进行排序，需要用到 sort_values()方法

- 在sort_values后的括号中指明要排序的列名，以及升序还是降序排列。

**语法**



```
df.sort_values(by = ["col1"], ascending = False)
```

- 上面代码表示df表按照 col1 列进行排序，ascending = False表示按照 col1 列进行降序排列。
- ascending 参数默认值为 True，表示升序排列。
- 所以，如果是要根据 col1 列进行升序排序，则可以只指明列名，不需要额外声明排序方式。


````python
# 导入依赖
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers.csv")

# df.sort_values(by = ["Age"])
df.sort_values(by = ["Age"], ascending=False)
````

### 按照有缺失值的列进行排序

- 当待排序的列中有缺失值时，可以通过设置   na_position 参数对缺失值的显示位置进行设置，默认参数值为 last ，可以不写，表示将缺失值显示在最后。

````python
# 导入依赖
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers_NAN.csv")

# 通过设置na_position参数将缺失值显示在最前面。
df.sort_values(by = ["Age"], na_position= "first")
````

### 按照多列数值进行排序

- 按照多列数值排序是指同时依据多列数据进行升序、降序排列，当第一列出现重复值时按照第二列进行排序，当第二列出现重复值时按照第三列进行排序

- 在 Python 中实现按照多列进行排序，用到的方法同样是 sort_values()，只要在sort-values后的括号中以列表的形式指明要排序的多列列名及每列的排序方式即可。



````python
# 导入依赖
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers_NAN.csv")

# 输出行索引index和列索引columns
print('列索引:{}'.format(df.columns))
print('行索引:{}'.format(df.index))
# 通过设置na_position参数将缺失值显示在最前面
df.sort_values(by = ["Age", "CustomerID"], ascending = [True, False])
# 上面代码表示df表先按照"Age"列进行升序排列，当Age列遇到重复时，
# 再按照CustomerID列进行降序排列。对于表df我们依旧先按照"Age"升序排列，
# 当遇到重复的销售ID时，再按成交时间降序排列
````

## 3.数值排名

- 数值排名和数值排序是相对应的，排名会新增一列，这一列用来存放数据的排名情况，排名是从1开始的。

- 在Python中对数值进行排名，需要用到rank()方法。rank()方法主要有两个参数，一个是ascending，用来指明升序排列还是降序排列，默认为升序排列，和Excel中order的意思一致；另一个是method，用来指明待排列值有重复值时的处理情况。

**method**

- average: 当待排名的数值有重复值时，返回重复值的平均排名。

- first: 按值在素有的待排列数据找那个出现的相互顺序排名

- min: 当待排名的数值有重复值时， RANK.EQ返回重复值的最佳排名。

- max: 与min相反，取重复值对应的最大排名

````python
# 导入依赖
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/BESTROE/dataset/Mall_Customers_NAN.csv")

print(df["Age"].rank(ascending = True, method = 'average'))

print(df["Age"].rank(ascending = True, method = 'first'))

print(df["Age"].rank(ascending = True, method = 'min'))

print(df["Age"].rank(ascending = True, method = 'max'))
````

## 6.区间切分

- 区间切分就是将一系列数值分成若干份，比如现在有10个人，你要根据这10个人的年龄将他们分为三组，这个切分过程就称为区间切分。

- 在Python中对区间切分利用的是cut()方法，cut()方法有一个参数bins用来指明切分区间。

- 与cut()方法类似的还有qcut()方法，qcut()方法不需要事先指明切分区间，只需要指明切分个数，即你要把待切分数据切成几份，然后它就会根据待切分数据的情况，将数据切分成事先指定的份数，依据的原则就是每个组里面的数据个数尽可能相等。

- 在数据分布比较均匀的情况下，cut()方法和 qcut()方法得到的区间基本一致，当数据分布不均匀，即方差比较大时，两者得到的区间的偏差就会比较大。

````python
# cut()方法的切分结果是几个左开右闭的区间
pd.cut(df['Age'], bins=[0,20,28,35])
````

````python
pd.qcut(df['Age'], 3)
````

上面这种由很多列转换为很多行的过程，就是宽表转换为长表的过程，这种转换过程是有前提的，那就是：
- **需要有公共列**

**实现方法**

- stack()

- melt()

````python
# stack()在将表格型数据转为树形数据时，是在保持行索引不变的前提下，将列索引也变成行索引。
# 新建表型数据表

df = pd.DataFrame([['Apple', '苹果', '5000', '5050', '5050', '5050'],
                  ['Google', '谷歌', '5000', '5050', '5050', '5050'],
                  ['Facebook', '脸书', '5000', '5050', '5050', '5050']],
                  columns = ['Company','Name','Sale2013','Sale2014','Sale2015','Sale2016'])
df
````

````python
df.set_index(["Company", "Name"])
````

````python
df.set_index(["Company", "Name"]).stack()
````

````python
# 需要先将 Company和Nmae先设置成索引，然后调用stack()方法，将列索引也转换成行索引，最后利用reset_index()方法进行索引重置
print(df)
df.set_index(["Company", "Name"]).stack().reset_index()
````

- melt中的id_vars参数用于指明宽表转换到长表时保持不变的列，

- var_name参数表示原来的列索引转化为“行索引”以后对应的列名，- value_name表示新索引对应的值的列名。

````python
# melt()方法实现上述任务
print(df)
df.melt(id_vars= ["Company", "Name"], var_name= "Year", value_name= "Sale")
````

### 长表准换成宽表

- 将长表转化为宽表就是宽表转化为长表的逆过程。

````python
df1 = df.melt(id_vars= ["Company", "Name"], var_name= "Year", value_name= "Sale")
print(df1)
df1.pivot_table(index = ["Company", "Name"], columns = ["Year"], values= ["Sale"], aggfunc='first')
````

## 11.apply()与 applymap()

- map()函数是对一个序列中的所有元素执行相同的函数操作。

- 在DataFrame中与map()函数类似的函数有两个，一个是apply()函数，另一个是applymap()函数。函数apply()和applymap()都需要与匿名函数lambda结合使用。

- apply()函数主要用于对DataFrame中的某一column或row中的元素执行相同的函数操作。

- applymap()函数用于对DataFrame中的每一个元素执行相同的函数操作。

````python
df = pd.DataFrame([[1,2,3], [4,5,6], [7,8,9]], columns = ['C1', 'C2', 'C3'])
df
````

````python
df['C1'].apply(lambda x: x-1)
````

````python
df.applymap(lambda x: x + 1)
````
